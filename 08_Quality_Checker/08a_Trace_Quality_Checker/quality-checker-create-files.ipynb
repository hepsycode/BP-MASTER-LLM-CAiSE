```python
import os
import pandas as pd
from collections import Counter

def load_file_lines(path):
    """Loads the content of a file and returns it as a list of lines without newline characters."""
    with open(path, 'r') as file:
        return [line.strip() for line in file.readlines()]

def count_words_by_position(lines):
    """
    Counts occurrences of words by individual positions (first, second, third, fourth),
    as well as combined occurrences of (second and third) and (second, third, fourth) words.
    Returns a list of Counter dictionaries for each position and combination.
    """
    position_counts = [Counter(), Counter(), Counter(), Counter(), Counter(), Counter()]
    
    for line in lines:
        words = line.split()
        # Count individual word positions
        for i, word in enumerate(words[:4]):  # Consider the first 4 words
            position_counts[i][word] += 1
        
        # Count combinations of second and third words
        if len(words) >= 3:
            position_counts[4][f"{words[1]} {words[2]}"] += 1
        
        # Count combinations of second, third, and fourth words
        if len(words) >= 4:
            position_counts[5][f"{words[1]} {words[2]} {words[3]}"] += 1
            
    return position_counts

def calculate_hallucination_metric(counts1, counts2):
    """
    Calculates the "hallucination" metric by taking the ratio of occurrences in counts1
    relative to counts2 for each word position and combination. If a word combination is 
    in counts1 but not in counts2, its occurrence count from counts1 is used directly.
    """
    hallucination_ratios = []
    for i in range(6):  # Loop over each position and combination (first 4 positions + 2 combinations)
        ratio_counter = {}
        for word, count1 in counts1[i].items():
            count2 = counts2[i].get(word, 0)
            if count2 == 0:
                ratio_counter[word] = (count1, count2, count2)  # Store occurrences and hallucination value as count1
            else:
                ratio_counter[word] = (count1, count2, count2 / count1)  # Store occurrences and hallucination ratio
        hallucination_ratios.append(ratio_counter)
    return hallucination_ratios

def compare_directories(dir1, dir2):
    """
    Compares files in the two specified directories for hallucination metrics.
    Only files with the same name in both directories are compared.
    """
    results_by_position = {i: [] for i in range(6)}  # Dictionary to store results for each position
    
    for filename in os.listdir(dir1):
        path1 = os.path.join(dir1, filename)
        path2 = os.path.join(dir2, filename)
        
        if os.path.isfile(path1) and os.path.isfile(path2):
            # Load lines from both files
            lines1 = load_file_lines(path1)
            lines2 = load_file_lines(path2)
            
            # Count word occurrences by position and combination in both files
            counts1 = count_words_by_position(lines1)
            counts2 = count_words_by_position(lines2)
            
            # Calculate hallucination metric for each position and combination
            hallucination_metric = calculate_hallucination_metric(counts1, counts2)
            
            # Append results for each position to the respective lists in results_by_position
            for pos, ratios in enumerate(hallucination_metric):
                for word, (occ1, occ2, ratio) in ratios.items():
                    results_by_position[pos].append({
                        "filename": filename,
                        "word_combination": word,
                        "occurrences_dir1": occ1,
                        "occurrences_dir2": occ2,
                        "hallucination_ratio": ratio
                    })

    # Convert each position's results into DataFrames
    df_results = {
        "First Word": pd.DataFrame(results_by_position[0]),
        "Second Word": pd.DataFrame(results_by_position[1]),
        "Third Word": pd.DataFrame(results_by_position[2]),
        "Fourth Word": pd.DataFrame(results_by_position[3]),
        "Second & Third Words": pd.DataFrame(results_by_position[4]),
        "Second, Third & Fourth Words": pd.DataFrame(results_by_position[5]),
    }
    
    return df_results

# Usage example
pathReal = str(os.getcwd()) + '\BPMN-Designer\XES-MORGAN-HD' + '\\'
pathLLM = str(os.getcwd()) + '\BPMN-Designer\XES-MORGAN-HD-LLM-GPT4o' + '\\'

# 'XES-MORGAN-HD-LLM-GPT3.5', 'XES-MORGAN-HD-LLM-GPT4', 'XES-MORGAN-HD-LLM-GPT4o', 'XES-MORGAN-HD-LLM-LLAMA3-70B'
# 'XES-MORGAN-MIWG-LLM-GPT3.5', 'XES-MORGAN-MIWG-LLM-GPT4', 'XES-MORGAN-MIWG-LLM-GPT4o', 'XES-MORGAN-MIWG-LLM-LLAMA3-70B'

pathFileNameResCSV = 'Results/results_RQ1/BPMN-Designer/Hallucination-HD-GPT4o.csv'

# 'Hallucination-HD-GPT35', 'Hallucination-HD-GPT4', 'Hallucination-HD-GPT4o', 'Hallucination-HD-Llama3-70B'
# 'Hallucination-MIWG-GPT35', 'Hallucination-MIWG-GPT4', 'Hallucination-MIWG-GPT4o', 'Hallucination-MIWG-Llama3-70B'

# Usage example
hallucination_results = compare_directories(pathReal, pathLLM)

# Save the hallucination metrics to an Excel file with each position in a separate sheet
pathFileNameRes = 'Results/results_RQ1/BPMN-Designer/Hallucination/Hallucination-HD-GPT4o.xlsx'
with pd.ExcelWriter(pathFileNameRes) as writer:
    for sheet_name, df in hallucination_results.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

# Save each position's results as a separate CSV file
for sheet_name, df in hallucination_results.items():
    csv_path = f"Results/results_RQ1/BPMN-Designer/Hallucination/Hallucination-HD-GPT4o_{sheet_name.replace(' ', '_').lower()}.csv"
    df.to_csv(csv_path, index=False)
    print(f"Saved {sheet_name} metrics to {csv_path}")

# Displaying the DataFrame results for verification (optional)
for sheet_name, df in hallucination_results.items():
    print(f"\n{sheet_name}:\n", df)
```
